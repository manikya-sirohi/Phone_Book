```python
import json
```

```python
class My_Phone_Book:
    def __init__ (self, filepath):
        self.filepath = filepath
    def load_data(self):
        try:
            with open(self.filepath, "r") as file:
                return json.load(file)
        except Exception as e:
            print(e)
            return {}
    def save_data(self,data):
        try:
            with open(self.filepath, "w") as file:
                json.dump(data, file, indent=4)
            print("Update was successful")
        except Exception as e:
            print(e)
    def add(self,number, name):
        data = self.load_data()
        data[name] = number
        self.save_data(data)
        # record.seek(0) # yee vaalaaa firse pointer ko starting mein karne ke liye h 
        # record.truncate()
        # json.dump(record_data,record,indent=4)
    def remove_entry(self,name):
        try:
            data = self.load_data()
            del data[name]
            # record.seek(0) 
            # record.truncate()
            self.save_data(data)
            print("Contact Removed Successfully")
        except Exception as e:
            print(e)
    def display(self):
        try:
            data = self.load_data()
            print(data)
        except Exception as e:
            print(e)
    def search(self,name):
        try:
            data = self.load_data()
            if(data.get(name)):
                print("The number associated with the name", name,"is",data.get(name))
            else:
                print("The required entry is not found")
        except Exception as e:
            print(e)
```

```python
def function_match(value):
    record = My_Phone_Book(r"D:\ANUDIP\Phone_Book\Data\Data.json")

    match value:
        case 1:
            record.display()
        case 2:
            name = input("Enter Name : ")
            number = input("Enter Number : ")
            record.add(number,name)
        case 3:
            name = input("Enter Name : ")
            record.remove_entry(name)
        case 4:  
            name = input("Enter Name : ")
            record.search(name)
        case 5:
            print("Exiting...")

        case _:
            print("Invalid choice")
```
